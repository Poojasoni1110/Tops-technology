19. Question: Create a random 4x4 matrix and use NumPy to compute its
eigenvalues and eigenvectors. Verify the eigenvalues by reconstructing the
matrix.

```python
import numpy as np

# Create random 4x4 matrix
np.random.seed(42)
A = np.random.rand(4, 4)

# Eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Reconstruct matrix
Lambda = np.diag(eigenvalues)
A_reconstructed = eigenvectors @ Lambda @ np.linalg.inv(eigenvectors)

# Verify
print("Original Matrix:\n", A)
print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)
print("\nReconstructed Matrix:\n", A_reconstructed)
print("\nVerification:", np.allclose(A, A_reconstructed))
```

21. Question: Perform matrix multiplication of two 2D arrays using np.dot ()
and @. Compare the results and performance of both methods using a large
dataset.

```python
import numpy as np
import time

# Step 1: Create two large random 2D arrays
a = np.random.randint(1, 10, (1000, 1000))
b = np.random.randint(1, 10, (1000, 1000))

# Step 2: Matrix multiplication using np.dot()
start_time = time.time()
result_dot = np.dot(a, b)
time_dot = time.time() - start_time
print("Time using np.dot():", time_dot)

# Step 3: Matrix multiplication using @ operator
start_time = time.time()
result_at = a @ b
time_at = time.time() - start_time
print("Time using @ operator:", time_at)

# Step 4: Verify both results are same
print("Results are equal:", np.array_equal(result_dot, result_at))
```
